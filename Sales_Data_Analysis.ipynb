```python
#Add and calculate a new Calculated_Date column
df['Calculated_Date'] = df[['Year', 'Month', 'Day']].apply(lambda x: '{}-{}-{}'.format(x[0], x[1], x[2]), axis=1)

df['Calculated_Date'].head()
```

```python
#Parse your Calculated_Date column into a datetime object
df['Calculated_Date'] = pd.to_datetime(df['Calculated_Date'])

df['Calculated_Date'].head()
```

```python
#How did sales evolve through the years?
#Show a line plot using Calculated_Date column as the x-axis and the count of sales as the y-axis.
df['Calculated_Date'].value_counts().plot(kind='line', figsize=(15,6))
```

```python
#What is the mean Order_Quantity of orders with more than 10K in revenue?
condition= df['Revenue']>10_000
df.loc[condition, 'Order_Quantity'].mean().round(2)
```

```python
#What is the mean Order_Quantity of orders with less than 10K in revenue?
condition_less_than_10K= df['Revenue']<10_000
df.loc[condition_less_than_10K,'Order_Quantity'].mean().round(2)
```
