```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"/content/sample_data/Ice Cream Ratings.csv")
df= df.set_index("Date")
df
```

```python
df.plot()
```

```python
df.plot(kind='line')
```

```python
df.plot(kind='line', subplots=True)
```

```python
df.plot(kind='line', title='Ice Cream Ratings')
```

```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel="Days", ylabel= "Scores")
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar', stacked= True)
```

```python
df['Flavor Rating'].plot(kind='bar')
```

```python
df.plot.barh(stacked= True)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating',s=100,c='red')
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
print(plt.style.available)
plt.style.use('classic')
df.plot.pie(y="Flavor Rating", figsize= (10,6))
```
