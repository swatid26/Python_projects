```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['SafetyIncidents']
  ys = series['count']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = Safety_Incident_count.sort_values('SafetyIncidents', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('SafetyIncidents')
_ = plt.ylabel('count')
```

```python
from sklearn.preprocessing import StandardScaler

```

```python
# Separate features and target variable
X = df.drop('DefectStatus', axis=1)
y = df['DefectStatus']

```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Initialize the model
model = RandomForestClassifier(random_state=42)

```

```python
# Train the model
model.fit(X_train, y_train)

```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python

# Evaluate the model
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
```

```python
# Feature importance
importances = model.feature_importances_

```

```python
import numpy as np
indices = np.argsort(importances)[::-1]
features = X.columns


```

```python
# Plot the feature importances
plt.figure(figsize=(12, 8))
plt.title('Feature Importances')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), features[indices], rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 10))  # Adjust size if needed
plt.title('Feature Importances')

# Pie chart
plt.pie(importances[indices], labels=features[indices], autopct='%1.1f%%', startangle=140)
```

```python
import numpy as np
from sklearn.svm import OneClassSVM

# Initialize the One-Class SVM model
oc_svm = OneClassSVM(gamma='auto', nu=0.05)  # nu is the upper bound on the fraction of margin errors

# Fit the model
oc_svm.fit(X_scaled)

# Predict anomalies
anomalies = oc_svm.predict(X_scaled)

# Convert anomaly labels: -1 for anomalies, 1 for normal
anomalies = np.where(anomalies == -1, 1, 0)

# Add anomalies to the DataFrame
df['Anomaly'] = anomalies

# Display results
print(df[df['Anomaly'] == 1])  # Print rows identified as anomalies
```

```python
# Initialize and fit the One-Class SVM model
oc_svm = OneClassSVM(gamma='auto', nu=0.05)
oc_svm.fit(X_scaled)

# Predict anomalies
anomalies = oc_svm.predict(X_scaled)

# Convert anomaly labels: -1 for anomalies, 1 for normal
anomalies = np.where(anomalies == -1, 1, 0)

# Add anomalies to the DataFrame
df['Anomaly'] = anomalies

# Remove anomalies from the dataset
X_cleaned = X[df['Anomaly'] == 0]
y_cleaned = y[df['Anomaly'] == 0]

# Standardize the cleaned features
X_cleaned_scaled = scaler.transform(X_cleaned)

# Split the cleaned data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_cleaned_scaled, y_cleaned, test_size=0.2, random_state=42)

# Train a RandomForestClassifier on the cleaned data
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = model.predict(X_test)
print(f'Accuracy on cleaned data: {accuracy_score(y_test, y_pred)}')
```
